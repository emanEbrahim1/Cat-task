```python
#Define a function to calculate IQR and apply it on the price 
def find_iqr(x):
    #q75, q25 = np.percentile(x, [75 ,25])
    #iqr = q75 - q25
  return np.subtract(*np.percentile(x, [75, 25]))

data[['price']].apply(find_iqr)
```

```python
#Q2 : build the following DataFrame 

datas = [[78,84,86],[85,94,97],[96,86,96],[80,83,72],[86,86,83]]
daf = pd.DataFrame(datas,
                  columns=['X', 'Y', 'Z'])
daf
```

```python
#Q3 : what is the difference between loc and iloc ? show that
#     by applying the two methods on the previous DataFrame . 

# Two used for slicing data or select data from datafram 
#loc()we have to pass the name of the row or column which we want to select(based on labels not index)
#can accept the boolean data unlike iloc()
#Many operations can be performed using the loc()
display(daf.loc[(daf.X > 86) & (daf.Z > 72)]) 
```

```python
#iloc() we have to pass integer index in the method to select specific row/column (based on index)
#does not accept the boolean data
display(daf.iloc[[0, 2, 4]]) 
display(daf.iloc[2:4 , 1 : ]) 
```
